## Boolean Indexing


Select rows from `df` where the value in column B is greater than or equal to 8.


```python
#your code here
df[df["B"] >= 8]
```

Select rows from df where the value in column B is greater than or equal to 8 and the value in column D is equal to "a".

```python
#your code here
df[(df["B"] >= 8) & (df["D"] == "a")]
```
